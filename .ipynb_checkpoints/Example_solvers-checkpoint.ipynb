# Simple Cantilever Beam Example

## Required Imports

```python
from fem2d.meshers import Mesh, Q8Mesh
from fem2d.solvers import FEMSolvers
from fem2d.postprocessing import Plotting
```

## Problem Set Up

```python
E = 210000 #youngs modulous
v = 0.3 #possions ratio
t = 1 #thickness
plane = 1 #plane strain or plain stress
ELtype = 'Q4' #Element type

Beam = Mesh(E, v, t, plane, ELtype) #Create mesh object

el_num = 20
Length = 10
thickness = 0.5
Load = -25

Beam.SimpleBeam(el_num, Length, thickness, Load) #Create a simple cantilever beam
```

## View Created Mesh

```python
#All plotting functions return the ax object.
ax = Plotting.InitialMesh(Beam)
# Plotting.ProblemDiagram(Beam) <- this will include node numbers.
```

## Simulate Problem

```python
# solvers return the displacement vector
U = FEMSolvers.LinearSolver(Beam)
ax = Plotting.Overlay(Beam)
```

### The same mesh object can be used with all the solvers

```python
U = FEMSolvers.NonLinearSolver(Beam, LoadSteps=3, tol = 1e-2)
ax = Plotting.Overlay(Beam, steps=1)
ax = Plotting.LoadPath(Beam)
```

```python
U = FEMSolvers.ArcLengthSolver(Beam, ArcLength=2, TotalArcLength=6, tol = 1e-2)
ax = Plotting.Overlay(Beam, steps=1)
```

## View Results

### The results are stored in the mesh. attributes

```python
Beam.U
Beam.LoadValues
```
